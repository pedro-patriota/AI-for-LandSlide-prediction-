```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.inspection import DecisionBoundaryDisplay

```

```python
df = pd.read_csv("new_df.csv")
df.drop(['em_recife', 'processo_numero', 'solicitacao_data', 'solicitacao_hora', 'solicitacao_descricao'],inplace=True, axis=1)
df.dropna(inplace=True)
```

```python
def label_ground_int(row):
    match row['label']:
        case "LVd2":
            return 0 
        case "LVd4":
            return 1 
        case "A2":
            return 2 
        case "URBANO":
            return 3 
        case "PV1":
            return 4 
        case "HP":
            return 5 
        case "AGUA":
            return 6 
        case "ILHA":
            return 7 
```

```python
reg = LogisticRegression(solver='liblinear',C=1e5)
df['label'].unique()
df['label_ground_int'] = df.apply(lambda row: label_ground_int(row), axis=1)
```

```python
train, test = train_test_split(df, test_size=0.5)
y = train['confirmado']
x = train[['altitude', 'rain_day', 'rain_hour', 'label_ground_int']]
x_test = test[['altitude', 'rain_day', 'rain_hour', 'label_ground_int']]
y_test = test['confirmado']

print(x.info())


reg = reg.fit(x, y)
y_pred = reg.predict(x_test )

print(reg.classes_)
print(reg.score(x_test, y_test))
print("-------")
print(y_pred)
prob = reg.predict_proba(x_test)
print(prob)
reg.score(x, y) 

```

```python
x = x.dropna(thresh=2, inplace=True)
print(x.info())

_, ax = plt.subplots(figsize=(4, 3))
DecisionBoundaryDisplay.from_estimator(
    reg,
    x,
    cmap=plt.cm.Paired,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    xlabel="Sepal length",
    ylabel="Sepal width",
    eps=0.5,
)

# Plot also the training points
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolors="k", cmap=plt.cm.Paired)


plt.xticks(())
plt.yticks(())

plt.show()
```
